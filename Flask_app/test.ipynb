```python
import os
import pandas as pd
from flask import Flask, render_template, redirect
from flask_sqlalchemy import SQLAlchemy
from sqlalchemy import create_engine


engine = create_engine(os.getenv("postgres_uri"))


```

```python
import plotly.graph_objects as go
import plotly
import numpy as np
```

```python

query = '''select * from mentions_and_price'''
price_mentions_data = pd.read_sql(query, engine)

price_mentions_data
```

```python
price_mentions_data.rename(columns={'Unnamed: 1':"Ticker"}, inplace=True)
```

```python
new_df = price_mentions_data.sort_values(by=['Ticker','date'])
```

```python
new_df
```

```python
new_df['Volume'] = new_df['Volume'].fillna(value=0)
```

```python
new_df = new_df.fillna(method='ffill')
```

```python
df = new_df
```

```python
gme = df[df['Ticker'] == "GME"].sort_values(by=['date'])
```

```python

trace1 = {
    'x': gme.date,
    
    'open': gme.Open,
    'close': gme.Close,
    'high': gme.High,
    'low': gme.Low,
    'type': 'candlestick',
    'name': 'GME',
    'showlegend': True
}
```

```python
data = [trace1]
# Config graph layout
layout = go.Layout({
    'title': {
        'text': 'GameStop(GME) Moving Averages',
        'font': {
            'size': 15
        }
    }
})
```

```python
fig = go.Figure(data, layout)
fig.show()
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter(x=gme.date, y=gme.Close,
                    mode='lines+markers',
                    name='lines'))
fig.add_trace(go.Bar(x=gme.date, y=gme.Open,
                    ))
fig.show()
```
