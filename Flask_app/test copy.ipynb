```python
import os
import pandas as pd
from flask import Flask, render_template, redirect
from flask_sqlalchemy import SQLAlchemy
from sqlalchemy import create_engine


engine = create_engine(os.getenv("postgres_uri"))


```

```python
import plotly.graph_objects as go
import plotly
```

```python

query = '''select * from mentions_and_price'''
price_mentions_data = pd.read_sql(query, engine)

price_mentions_data
```

```python
price_mentions_data.rename(columns={'Unnamed: 1':"Ticker"}, inplace=True)
```

```python
new_df = price_mentions_data.set_index('Ticker').sort_index()
```

```python
new_df
```

```python
new_df['Volume'] = new_df['Volume'].fillna(value=0)
```

```python
new_df = new_df.fillna(method='ffill')
```

```python
df = new_df
```

```python

trace1 = {
    'x': df.index,
    'open': df.Open,
    'close': df.Close,
    'high': df.High,
    'low': df.Low,
    'type': 'candlestick',
    'name': 'MSFT',
    'showlegend': True
}
```

```python
data = [trace1]
# Config graph layout
layout = go.Layout({
    'title': {
        'text': 'Microsoft(MSFT) Moving Averages',
        'font': {
            'size': 15
        }
    }
})
```

```python
fig = go.Figure(data, layout)
fig.show()
```
