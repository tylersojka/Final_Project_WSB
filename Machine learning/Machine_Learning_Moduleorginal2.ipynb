```python
from path import Path
import pandas as pd
import numpy as np
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import yfinance as yf
```

```python
combo_multi_path = "Resources/combo_count_multiindex.csv"
ticker_path = "Resources/final_ticker_list.csv"
```

```python
count_df = pd.read_csv(combo_multi_path, header=[0,1], index_col=0)
ticker_list = pd.read_csv(ticker_path)
```

```python
ticker_list = ticker_list["Tickers"].to_list()
```

```python
count_df.head()
```

```python
df = pd.DataFrame(yf.download(ticker_list, start="2020-12-08", end="2021-02-05", interval="1d", group_by='tickers', index=False))
```

```python
amc_count = pd.DataFrame(count_df["AMC"])
amc_price = df["AMC"]
```

```python
# enter combined AMC df info
amc_result = amc_count.join(amc_price)
amc_result
```

```python
amc_result["Volume"]= amc_result["Volume"].fillna(0)
```

```python
amc_result.fillna(method="ffill",inplace=True)
```

```python
#add "Day","Month","Year"
amc_result
```

```python
y = amc_result['Volume']
X = amc_result.drop([ 'High'], axis=1)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 1)
```

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(solver='1bfgs',
                                max_iter=200,
                                random_state=1)
```

```python
classifier.fit(X_train, y_train)
```

```python
y_pred=classifier.predict(X_test)
results=pd.DataFrame({"Prediction": y_pred, "Actual": y_test}).reset_index(drop=True)
results.head()
```
